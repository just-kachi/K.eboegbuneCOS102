```python
import tkinter as tk
from tkinter import messagebox, filedialog
import csv

def load_data_from_csv(file_path):
    employees = {}
    try:
        with open(file_path, 'r', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader:
                name = f"{row['FIRST NAME']} {row['SURNAME']}"
                department = row['DEPARTMENT']
                employees[name] = department
        return employees
    except FileNotFoundError:
        messagebox.showerror("Error", f"File not found: {file_path}")
    except Exception as e:
        messagebox.showerror("Error", f"Error loading data from file: {str(e)}")
    return {}

def check_employee():
    employee_name = name_entry.get().strip().title()
    department = department_entry.get().strip().title()

    if employee_name in employees and employees[employee_name] == department:
        welcome_message = f"Welcome {employee_name} to {department} department!\n"
        messagebox.showinfo("Welcome", welcome_message)
        
        department_members = [name for name, dept in employees.items() if dept == department and name != employee_name]
        if department_members:
            members_message = f"Other members of {department} department: {', '.join(department_members)}"
            messagebox.showinfo("Department Members", members_message)
        else:
            messagebox.showinfo("Department Members", f"There are no other members in {department} department.")

    else:
        messagebox.showinfo("Not Employed", f"{employee_name} is not employed in {department} department.")

def open_file():
    file_path = r"C:\Users\USER\Documents\K.eboegbuneCOS102\Week 5\GIG-logistics.csv"
    try:
        global employees
        employees = load_data_from_csv(file_path)
        if employees:
            messagebox.showinfo("Success", "Data loaded successfully from file.")
        else:
            messagebox.showerror("Error", "No data loaded from file.")
    except Exception as e:
        messagebox.showerror("Error", f"Error loading data from file: {str(e)}")

# Create main window
root = tk.Tk()
root.title("Employee Verification")
root.geometry("400x250")

# Create buttons for loading file and checking employee
load_button = tk.Button(root, text="Load Data from File", command=open_file)
load_button.pack(pady=5)

# Create name label and entry
name_label = tk.Label(root, text="Enter Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Create department label and entry
department_label = tk.Label(root, text="Enter Department:")
department_label.pack()
department_entry = tk.Entry(root)
department_entry.pack()

# Create submit button
submit_button = tk.Button(root, text="Check Employee", command=check_employee)
submit_button.pack(pady=5)

# Initialize employees dictionary
employees = {}

# Run the main event loop
root.mainloop()
```
