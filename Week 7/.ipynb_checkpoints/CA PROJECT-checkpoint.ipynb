```python
import tkinter as tk
from tkinter import messagebox

menu = [
    ("RICE/PASTA", [
        ("Jollof Rice", 350),
        ("Coconut Fried Rice", 350),
        ("Jollof Spaghetti", 350)
    ]),
    ("SIDE DISHES", [
        ("Savoury Beans", 350),
        ("Roasted Sweet Potatoes", 300),
        ("Fried Plantain", 150),
        ("Mixed Vegetable Salad", 150),
        ("Boiled Yam", 150)
    ]),
    ("PROTEINS", [
        ("Sweet Chill Chicken", 1100),
        ("Grilled Chicken Wings", 400),
        ("Fried Beef", 400),
        ("Fried Fish", 500),
        ("Boiled Egg", 200),
        ("Sauteed Sausages", 200)
    ]),
     ("SOUPS & SWALLOWS", [
        ("Eba", 100),
        ("Poundo Yam", 100),
        ("Semo", 100),
        ("Atama Soup", 450),
        ("Egusi Soup", 400)
    ]),
    ("BEVERAGES", [
        ("Water", 200),
        ("Glass Drink(35cl)", 150),
        ("PET Drink(35cl)", 300),
        ("PET Drink(50cl)", 350),
        ("Glass/Canned Malt", 500),
        ("Fresh Yo", 600),
        ("Pineapple Juice", 350),
        ("Mango Juice", 350),
        ("Zobo Drink", 350)
    ]),
]

def display_menu():
    menu_window = tk.Toplevel(root)
    menu_window.title("Menu")
    
    for category_tuple in menu:
        category_name = category_tuple[0]
        items_list = category_tuple[1]

        category_label = tk.Label(menu_window, text=category_name)
        category_label.pack()

        for item, price in items_list:
            item_label = tk.Label(menu_window, text=f"{item}: {price}")
            item_label.pack()

menu_items = {
    "Jollof Rice": 350,
    "Coconut Fried Rice": 350,
    "Jollof Spaghetti": 350,
    "Savoury Beans": 350,
    "Roasted Sweet Potatoes": 300,
    "Fried Plantain": 150,
    "Mixed Vegetable Salad": 150,
    "Boiled Yam": 150,
    "Sweet Chill Chicken": 1100,
    "Grilled Chicken Wings": 400,
    "Fried Beef": 400,
    "Fried Fish": 500,
    "Boiled Egg": 200,
    "Sauteed Sausages": 200,
    "Eba": 100,
    "Poundo Yam": 100,
    "Semo": 100,
    "Atama Soup": 450,
    "Egusi Soup": 400,
    "Water": 200,
    "Glass Drink(35cl)": 150,
    "PET Drink(35cl)": 300,
    "PET Drink(50cl)": 350,
    "Glass/Canned Malt": 500,
    "Fresh Yo": 600,
    "Pineapple Juice": 350,
    "Mango Juice": 350,
    "Zobo Drink": 350
}

selected_items = []

def add_to_order():
    item = total_order_var.get()
    quantity = quantity_var.get()

    if item and quantity:
        selected_items.append((item, int(quantity)))
        order_listbox.insert(tk.END, f"{item}: {quantity}")

def calculate_total():
    total_cost = sum(menu_items[item] * quantity for item, quantity in selected_items)
    discount = 0
    
    # Apply discounts based on the total cost
    if 1000 < total_cost < 2500:
        discount = total_cost * 0.1
    elif 2500 <= total_cost < 5000:
        discount = total_cost * 0.15
    elif total_cost >= 5000:
        discount = total_cost * 0.25
    
    final_cost = total_cost - discount
    
    cost_label.config(text=f"Total Cost: ₦{total_cost}\nDiscount: ₦{discount}\nFinal Cost: ₦{final_cost}")

def open_order_page():
    order_page = tk.Toplevel(root)
    order_page.title("Order Form")
    
    frame = tk.Frame(order_page)
    frame.pack(padx=10, pady=10)
    
    total_order_label = tk.Label(frame, text="Select Item:")
    total_order_label.grid(row=0, column=0, sticky="w")
    
    total_order_menu = tk.OptionMenu(frame, total_order_var, *menu_items.keys())
    total_order_menu.grid(row=0, column=1, padx=5, pady=5)
    
    quantity_label = tk.Label(frame, text="Quantity:")
    quantity_label.grid(row=1, column=0, sticky="w")
    
    quantity_entry = tk.Entry(frame, textvariable=quantity_var)
    quantity_entry.grid(row=1, column=1, padx=5, pady=5)
    
    add_button = tk.Button(frame, text="Add to Order", command=add_to_order)
    add_button.grid(row=2, columnspan=2, pady=10)
    
    global order_listbox
    order_listbox = tk.Listbox(order_page, selectmode=tk.MULTIPLE, width=40, height=10)
    order_listbox.pack(pady=10)
    
    calculate_button = tk.Button(order_page, text="Calculate Total", command=calculate_total)
    calculate_button.pack(pady=10)
    
    global cost_label
    cost_label = tk.Label(order_page, text="")
    cost_label.pack()

# GUI setup
root = tk.Tk()
root.title("PAU CAFETERIA MENU")

total_order_var = tk.StringVar(root)
quantity_var = tk.StringVar(root)

view_menu_button = tk.Button(root, text="View Menu", command=display_menu)
view_menu_button.pack(pady=20)

new_order_button = tk.Button(root, text="New Order", command=open_order_page)
new_order_button.pack(pady=20)

root.mainloop()
```
